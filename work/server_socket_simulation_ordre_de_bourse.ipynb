# -1) imports

```python
import socket
import random
import time
```

```python
def print_time():
    from datetime import datetime
    print("%s min %s s"%(datetime.now().minute,datetime.now().second))
```

```python
print_time()
```

# 0) configuration

```python
port = 12072
ip   = socket.gethostname()
```

```python
with open("port.param", "w") as f:
    f.write(str(port))
```

# 1) création d'une socket

```python
serversocket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
serversocket.bind((ip, port))
print_time()
print("listen on %s:%s"%(ip, port))
serversocket.listen(1)
```

# 2) attendre que spark se connecte

```python
(clientsocket, address) = serversocket.accept()
```

```python
print("Connection de %s :\n %s"%(address, clientsocket))
```
