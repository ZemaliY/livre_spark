# -1) imports

```python
import socket
import random
import time
```

# 0) configuration

```python
port = 12030
ip   = socket.gethostname()
```

# 1) crÃ©ation d'une socket

```python
serversocket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
serversocket.bind((ip, port))
serversocket.listen(1)
```

# 2) attendre que spark se connecte

```python
(clientsocket, address) = serversocket.accept()
```

```python
print("Connection de %s :\n %s"%(address, clientsocket))
```

# 3) envoie de donnÃ©es

```python
serversocket.setblocking(False) 
```

```python
nb_client = 1000
nb_achat  = 5
clients   = ["client_%s"%x for x in range(nb_client)]
achats    = [(random.choice(clients), random.randint(0, 100)) for x in range(nb_achat)]
```

```python
tps_attente = 1
nb_achat    = 50
for i in range(30):
    print(i)
    time.sleep(tps_attente)
    achats    = [(random.choice(clients), random.randint(0, 100)) for x in range(nb_achat)]
    for client, valeur in achats:
        to_send = '%s,%s\n'%(client, valeur)
        print("sent %s bytes"%clientsocket.send(to_send.encode()))
```

# Envoi boston

```python
from sklearn import datasets
import pandas as pd
```

```python
boston = datasets.load_boston()
x      = pd.DataFrame(boston.data  , columns=boston.feature_names )
y      = pd.DataFrame(boston.target, columns=["MEDV"]             )
boston = pd.concat([x, y], axis=1)
```

```python
nb_d_envois = 10
tps_attente = 3
for i in range(nb_d_envois):
    time.sleep(tps_attente)
    sep     = ";"
    to_send = sep.join(["%s"%x for x in random.choice(x.values)]) + "\n"
    clientsocket.send(to_send.encode())
```
