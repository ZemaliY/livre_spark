```python
# imports
import socket
from pyspark.sql import SparkSession
from pyspark.streaming import StreamingContext
def print_time():
    from datetime import datetime
    print("%s min %s s"%(datetime.now().minute,datetime.now().second))
# paramètrage
listen_to_ip   = socket.gethostname()
with open("port.param", "r") as f:
    port           = f.readline()
    listen_to_port = int(port)
print("Going to listen on %s:%s"%(listen_to_ip, listen_to_port))
nb_secondes    = 4
checkpoint_dir = "./checkpoint/"
print_time()
# Initialisation de spatrk
spark       = SparkSession.builder.getOrCreate()
sc          = spark.sparkContext
print_time()
# création du streaming context
ssc         = StreamingContext(sc, nb_secondes)
dstream     = ssc.socketTextStream(listen_to_ip, listen_to_port)
ssc.checkpoint(checkpoint_dir)
print_time()
# Fonction d'update:
def mise_a_jour(liste_de_nouvelles_valeurs, valeur_actuelle ):
    if valeur_actuelle is None:
        valeur_actuelle = 0
    return sum(liste_de_nouvelles_valeurs) +  valeur_actuelle
        
# création du Directed Acyclic Graph :
data            = dstream.map(lambda x: x.split(","))
clients_facture = data.map(lambda x: (x[0], float(x[1])) )
update_client   = clients_facture.updateStateByKey(mise_a_jour)
update_client.pprint()

# démarrage de l'écoute de la socket : 
print_time()
ssc.start()
```

```python
ssc.stop(False, True)
```
