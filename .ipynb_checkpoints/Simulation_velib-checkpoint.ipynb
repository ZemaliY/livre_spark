```python
source = ColumnDataSource(data=dict(
    x=[1, 2, 3, 4, 5],
    y=[2, 5, 8, 2, 7],
    desc=['A', 'b', 'C', 'd', 'E'],
))

TOOLTIPS = [
    ("index", "$index"),
    ("(x,y)", "($x, $y)"),
    ("desc", "@desc"),
]

p = figure(plot_width=400, plot_height=400, tooltips=TOOLTIPS,
           title="Mouse over the dots")

p.circle('x', 'y', size=20, source=source)

show(p)
```

```python
TOOLTIPS = [
    ("index", "$id_cycliste"),
    ("(x,y)", "($longitude, $lattitude)"),
]

p = figure(plot_width=400, plot_height=400, tooltips=TOOLTIPS,
           title="Mouse over the dots")

p.circle('longitude', 'lattitude', size=20, source=ColumnDataSource(df))

show(p)
```

```python
from bokeh.models import HoverTool
```

```python
# Create a ColumnDataSource from df: source
source = ColumnDataSource(df)

# Create the figure: p
p = figure()
p.circle('longitude', 'lattitude', source= source)# Add circle glyphs to the figure p
hover = HoverTool(tooltips=[("index", "@index"), ('Name', '@index')])
p.add_tools(hover)
show(p)
```
