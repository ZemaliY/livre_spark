```python
def gogo(cyclistes):
    import notebook_importer
    import math
    from Cycliste    import Cycliste
    cyclistes = list(cyclistes)
    une_semaine = 3600*24
    for timestamp in range(une_semaine):
        timestamp = timestamp * 60
        for cycliste in cyclistes:
            cycliste.avancer(timestamp)
            
        if timestamp%(3600*6 )==0:
            new_cyclistes = [Cycliste(Position.get_random_position(cycliste.ville.x_max,cycliste.ville.y_max), 
                                     Position.get_random_position(cycliste.ville.x_max, cycliste.ville.y_max), 
                                     cycliste.ville)
                                for i in range(math.ceil(len(cyclistes)/100))]
            cyclistes.extend(new_cyclistes)
            
    ville = cycliste.ville
    [x.flush_data() for x in ville.prestataire.velos]
    [x.flush_logs() for x in ville.prestataire.reseau]
```
